# Simulation Trajectory 

```python
wheel_radius = 2.7
wheel_base_half = 7.5
sonar_zero_distance = 13.8

init_x = 0.0
init_y = 0.0
init_angle = 0.0

x_cam_noise = (0.0, 49.0)
y_cam_noise = (0.0, 49.0)
gyro_noise = (0.0, math.radians(16.0))
sonar_normal_noise = (0.0, 4.0)
sonar_invalid_noise = (0.0, 1e+6)

robot = pd.read_csv("log_robot_2.csv", sep=";")

def simulate(robot):
    angles = []
    x = []
    y = []
    x.append(init_x)
    y.append(init_y)
    angles.append(init_angle)

    for i in range(0, len(robot.index) - 1):
        v = (math.radians(robot.iloc[i]["right_speed"]) + math.radians(robot.iloc[i]["left_speed"])) * wheel_radius / 2.0
        w = (math.radians(robot.iloc[i]["right_speed"]) - math.radians(robot.iloc[i]["left_speed"])) * wheel_radius / (2.0 * wheel_base_half)
        dt = (robot.iloc[i + 1]["time"] - robot.iloc[i]["time"])
        x.append(x[i] + v * dt * math.cos(angles[i] + w * dt))
        y.append((y[i] + v * dt * math.sin(angles[i] + w * dt)))
        angles.append(angles[i] + w * dt)

    for i in range(0, len(y) - 1):
        y[i] = -1 * y[i]
    
    return x, y

x, y = simulate(robot)

camera = pd.read_csv("log_camera_2.csv", sep=";")
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, y)
plt.plot()
```

# Gyro and sonar trajectory

```python
def gyro(robot):
    x, y = simulate(robot)

    for i in range(1, len(y) - 1):
        y[i] = robot.iloc[i]["distance"] * math.cos(math.radians(robot.iloc[i]["angle"])) - 13.8

    return x, y

x, y = gyro(robot)

plt.plot(x, y)
```

# All Together

```python
x, y = simulate(robot)
x1, y1 = gyro(robot)
plt.plot(camera[["x"]], camera[["y"]], "r-", x, y, "b-", x1, y1, "g-")
```

# Kalman Filter

```python

def predict(x, P, F, Q, control, dt, vl, vr):
    nx = np.dot(F, x) + control(x, dt, vl, vr)
    nP = np.dot(F, P).dot(F.T) + Q
    return nx, nP

def update(H, P, R, x, z):
    y = z - np.dot(H, x)
    A = inv(np.dot(H, P).dot(H.T) + R)
    K = np.dot(P, H.T).dot(A)
    nx = x + np.dot(K, y)
    nP = P - np.dot(K, H).dot(P)
    
    return nx, nP

def kalman(x, P, measures, R, Q, F, H, control):
    xs, Ps = [], []
    dt = 0.0
    for i, m in enumerate(measures):
        j = i if i < len(robot) else len(robot)- 1
        x, P = predict(x, P, F, Q, control, dt, robot.iloc[j]["left_speed"], robot.iloc[j]["right_speed"])
        print(x)
        x, P = update(H, P, R, x, m)
        xs.append(x)
        Ps.append(P)
        dt = (camera.iloc[i]["time"] - camera.iloc[i - 1]["time"])
    return np.array(xs), np.array(Ps)
```

```python
x = np.array([[0.], [0.]])
P = np.array([[0., 0.], [0., 0.]])
F = np.array([[1., 0], [0., 1.]])
Q = np.array([[1., 0], [0., 1.]])
H = np.array([[1., 0], [0., 1.]])
m = np.array([[[x, y]] for x, y in zip(camera["x"], camera["y"])])
R = np.array([[49., 0.], [0., 49.]])

last_angle = 0.0

def control(x, dt, vl, vr):
    global last_angle
    v = (math.radians(vr) + math.radians(vl)) * wheel_radius / 2.0
    w = (math.radians(vr) - math.radians(vl)) * wheel_radius / (2.0 * wheel_base_half)
    xs = x[0][0] + v * dt * math.cos(last_angle + w * dt)
    ys = x[1][0] + v * dt * math.sin(last_angle + w * dt)
    last_angle = last_angle + w * dt
    
    return np.array([[xs], [ys]])

xs, p = kalman(x, P, m, R, Q, F, H, control)
```

# Issues with Kalman Filter

Because Sonar Noise variance depends on the angle, and this is not a differentiable function, we cannot properly calculate noise matrix in this case. This makes kalman ineffective when working with such unstable sensors.
